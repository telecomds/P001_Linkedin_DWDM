```python
from selenium import webdriver
import time
import pandas as pd
```

```python
url = 'https://www.linkedin.com/jobs/search?keywords=DWDM&location=worldwide&geoId=&trk=public_jobs_jobs-search-bar_search-submit&position=1&pageNum=0'

wd = webdriver.Chrome(executable_path='C:\Windows\chromedriver.exe')
wd.get(url)
```

```python
# Extract the number of results for search
no_of_jobs_path = '/html/body/div[1]/div/main/div/h1/span[1]'
no_of_jobs = int(wd.find_element_by_xpath(no_of_jobs_path).get_attribute('innerText'))
no_of_jobs
```

```python
#Scroll down the webpage to display all jobs in query
i = 1
while i <= int(no_of_jobs/25)+1:
    print(i,end="")  #to check kernel is working
    wd.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    i = i + 1
    try:
        wd.find_element_by_xpath('/html/body/div[1]/div/main/section[2]/button').click()
        time.sleep(5)
    except:
        pass
        time.sleep(5)
        
```

```python
# Obtain a list of jobs found in search
job_lists = wd.find_element_by_class_name('jobs-search__results-list')
jobs = job_lists.find_elements_by_tag_name('li') # return a list
```

```python
# We scrap webpage to obtain data
job_id= []
job_title = []
company_name = []
location = []
date = []
job_link = []
i = 1
for job in jobs:
    print(i,end="")  #To check kernel is working
    job_id0 = i
    job_id.append(job_id0)

    job_title0 = job.find_element_by_css_selector('h3').get_attribute('innerText')
    job_title.append(job_title0)

    company_name0 = job.find_element_by_css_selector('h4').get_attribute('innerText')
    company_name.append(company_name0)

    location0 = job.find_element_by_css_selector('[class="job-search-card__location"]').get_attribute('innerText')
    location.append(location0)
    
    date0 = job.find_element_by_css_selector('div>div>time').get_attribute('datetime')
    date.append(date0)

    job_link0 = job.find_element_by_css_selector('a').get_attribute('href')
    job_link.append(job_link0)
    
    i = i+1
```

```python
jd = []
seniority = []

i=1
for item in range(len(jobs)):
    print(i,end='')
    try:
        job_click_path = f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/div'
        job_click = job.find_element_by_xpath(job_click_path).click()
        time.sleep(5)
    except:
        job_click_path = f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/a/div[2]/h3'
        job_click = job.find_element_by_xpath(job_click_path).click()
        time.sleep(5)
 
    try:
        jd_path = '/html/body/div[1]/div/section/div[2]/section[2]'
        jd0 = job.find_element_by_xpath(jd_path).get_attribute('innerText')
        jd.append(jd0)
    except:
        jd.append('N/A')
        continue

    try:
        seniority_path = '/html/body/div[1]/div/section/div[2]/section[2]/ul/li[1]/span'
        seniority0 = job.find_element_by_xpath(seniority_path).get_attribute('innerText')
        seniority.append(seniority0)
    except:
        seniority.append('N/A')
        continue
        
    i = i+1
```

```python
#Check all vector have the same lenggh
print(len(date),len(company_name),len(job_title),len(location),len(jd),len(seniority),len(job_link))
```

```python
job_data = pd.DataFrame({'ID': job_id,
'Date': date,
'Company': company_name,
'Title': job_title,
'Location': location,
'Description': jd,
'Level': seniority,
#'Type': emp_type,
#'Function': job_func,
#'Industry': industries,
'Link': job_link
})
# cleaning description column
job_data['Description'] = job_data['Description'].str.replace('\n',' ')
```
